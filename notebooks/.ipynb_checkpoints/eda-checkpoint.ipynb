# Jointures

## Transactions & products

```python
# Utilisation de id_prod pour la clé pour une jointure externe à gauche (pour conserver nos transactions)
transactions_products = pd.merge(transactions, products, on="id_prod", how="left", indicator=True)
```

```python
# On va créer une fonction qui permet de tester si la jointure a bien été effectuée.
# On va créer un data frame temporaire pour rechercher une valeure qui n'est pas égale à "both" pour "_merge"
# On récupère les infos
# Si il y a des résultats différents de "both", ils seront affichés dans un head()

def test_merge(transactions_products):
    df_test_transactions_products = transactions_products[transactions_products["_merge"] != "both"]
    df_test_transactions_products.info()

    if(df_test_transactions_products.shape[0] > 0):
        display(df_test_transactions_products.head())

test_merge(transactions_products)
```

Pas mal de valeurs manquantes dans les colonnes "price" et "categ".

## Discrétisation de l'âge

Application de la loi de Sturges pour une répartition de l'âge. Il faut définir une valeur approximative pour le nombre "k" en fonction de la taille N de l'échantillon.

[Discrétiser sur combien de classes ?](https://magrit.hypotheses.org/288)

```python
# Vérification du nombre de clients qui servira en nombre d'observations (N)
clients = df_final["client_id"].nunique()
clients
```

```python
# Formule de Sturges-Huntsberger : k = 1 ° (10 / 3) * log(N)

result = 1 + ((10 / 3) * math.log10(clients))
print(
    "Le résultat suite à l'aide de la formule de Sturges-Huntsberger est {:,.2f}"
    .format(result))
```

```python
# On regarde maintenant les valeurs minimales et maximales pour l'âge
df_final[["age"]].describe()
```

- L'âge minimum est de 21 ans, le maximum est de 96
- Le résultat de la formule est de 14.
- N sera donc égal à cette valeur (14)

```python
min_age = 21
max_age = 96
n = 14

# Calcul de la tranche par intervalle
interval = (max_age - min_age) / n
round(interval,2)
```

```python
# Boucle pour définir automatiquement nos tranches d'âges qui seront attribuées
age = min_age

for elem in range(1, n):
    age = age + interval
    print(age)
```

```python
labels = [
    "21-26", "27-31", "32-37", "38-42", 
    "43-47", "48-53", "54-58", "59-63", 
    "64-69", "70-74", "75-79", "80-85", 
    "86-90", "90-96"
]

df_final["range_age"] = pd.cut(df_final["age"], 14, labels=labels)

df_final.sample(20)
```

### Transactions

```python
transactions = df_final.groupby(["year", "month"])["price"].sum()
transactions.index = pd.to_datetime([f"{y}-{m:02d}" for y, m in transactions.index])
transactions = transactions.reset_index()
transactions.columns = ["date", "price"]

plt.figure(figsize=(10, 4))
sns.lineplot(data=transactions, x="date", y="price", marker="o")
plt.title("Données de transactions du 03/2021 au 02/2023")
plt.xlabel("Date")
plt.ylabel("Somme des prix")

plt.tight_layout()
plt.show()
```

On remarque une baisse des ventes au mois d'octobre de l'année 2021. Analysons ça en détail.

#### Baisse d'octobre 2021

```python
october_2021 = df_final.loc[
    (df_final["year"] == 2021) & (df_final["month"] == 10),
    ["day", "price", "categ"]
]

october_2021 = october_2021.groupby(["day", "categ"]).agg({
    "price": ["sum", "count"]
}).reset_index()
```

```python
october_2021.columns = ["day", "categ", "ca", "volume"]

october_2021 = pd.pivot_table(october_2021, columns=["categ"], index=["day"]).reset_index()
october_2021.columns = ["day", "categ_0", "categ_1", "categ_2", "vol_0", "vol_1", "vol_2"]

october_2021
```

La baisse est donc lié à plusieurs "NaN" durant quasiment tout le mois d'octobre 2021 pour la catégorie 1 des produits. Il y a possiblement des données qui ont été perdues lors de l'export de la base de données vers le .csv des données.

```python
fig, axs = plt.subplots(3, 1, figsize=(12, 9), sharex=True)

for i, categ in enumerate([0, 1, 2]):
    axs[i].bar(october_2021["day"], october_2021[f"vol_{categ}"])
    axs[i].set_ylabel(f"Catégorie {categ}", fontsize=12)
    axs[i].set_title(f"Ventes journalières – Catégorie {categ}", fontsize=14)

axs[2].set_xlabel("Jour du mois d'octobre 2021", fontsize=12)
plt.suptitle("Analyse des volumes de ventes par jour et par catégorie – Octobre 2021", fontsize=16)
plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```
